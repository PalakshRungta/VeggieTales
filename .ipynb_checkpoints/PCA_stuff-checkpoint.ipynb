```python
# importing the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import seaborn as sns
from math import gcd
from PIL import Image
import cv2
from sklearn.decomposition import PCA
```

```python
#improve plotting resolution
%config InlineBackend.figure_format ='retina'
```

```python
# reads the apple jpg and displays it
img = cv2.cvtColor(cv2.imread('./processed_train/apple/Image_10.jpg'), cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
app10 = cv2.imread('./processed_train/apple/Image_10.jpg')
app_arr = cv2.split(app10) # splits in BGR order not RGB
```

```python
# displaying the blue channel
plt.title("Blue Channel")
plt.imshow(app_arr[0])
```

```python
# data of blue channel
blue_df = pd.DataFrame(data = app_arr[0])
blue_df
```

```python
# displaying the green channel
plt.title("Green Channel")
plt.imshow(app_arr[1])
```

```python
# data of green channel
green_df = pd.DataFrame(data = app_arr[1])
green_df
```

```python
# displaying the red channel
plt.title("Red Channel")
plt.imshow(app_arr[2])
```

```python
# data of red channel
red_df = pd.DataFrame(data = app_arr[2])
red_df
```

```python
# scale the data
df_blue = app_arr[0]/255
df_green = app_arr[1]/255
df_red = app_arr[2]/255
```

```python
# fit the data in PCA
pca_b = PCA(n_components=50)
pca_b.fit(blue_df)
trans_pca_b = pca_b.transform(blue_df)

pca_g = PCA(n_components=50)
pca_g.fit(green_df)
trans_pca_g = pca_g.transform(green_df)

pca_r = PCA(n_components=50)
pca_r.fit(red_df)
trans_pca_r = pca_r.transform(red_df)

# shape of tranformed data
print(trans_pca_b.shape)
print(trans_pca_r.shape)
print(trans_pca_g.shape)
```

```python
# bar plots to check variance ratio by each eigenvalue
fig = plt.figure(figsize = (15, 7.2)) 

fig.add_subplot(131)
plt.title("Blue Channel")
plt.ylabel('Variation explained')
plt.xlabel('Eigen Value')
plt.bar(list(range(1,51)),pca_b.explained_variance_ratio_)

fig.add_subplot(132)
plt.title("Green Channel")
plt.ylabel('Variation explained')
plt.xlabel('Eigen Value')
plt.bar(list(range(1,51)),pca_g.explained_variance_ratio_)

fig.add_subplot(133)
plt.title("Red Channel")
plt.ylabel('Variation explained')
plt.xlabel('Eigen Value')
plt.bar(list(range(1,51)),pca_r.explained_variance_ratio_)

plt.show()
```

```python
# reconstruct to validate and visualize
b_arr = pca_b.inverse_transform(trans_pca_b)
g_arr = pca_g.inverse_transform(trans_pca_g)
r_arr = pca_r.inverse_transform(trans_pca_r)
print(b_arr.shape, g_arr.shape, r_arr.shape)
```

```python
img_reduced= (cv2.merge((b_arr, g_arr, r_arr)))
print(img_reduced.shape)
```

```python
fig = plt.figure(figsize = (10, 7.2)) 
fig.add_subplot(121)
plt.title("Original Image")
plt.imshow(img)
fig.add_subplot(122)
plt.title("Reduced Image")
plt.imshow(img_reduced)
plt.show()
```

Apply PCA with original number of dimensions of blue to create the scree plot to see how well PCA captures the variance of the data.

```python
b_pca_1200 = PCA(n_components=1200)
b_pca_1200.fit(b_arr)

plt.grid()
plt.plot(np.cumsum(b_pca_1200.explained_variance_ratio_ * 100))
plt.xlabel('Number of components')
plt.ylabel('Explained variance')
plt.savefig('Scree plot.png')
```

```python
pca_50 = PCA(n_components=50)
b_pca_50_reduced = pca_50.fit_transform(b_arr)
# inverse_transform() method is used to decompress the reduced dataset back to 1200 dimensions
b_pca_50_recovered = pca_50.inverse_transform(b_pca_50_reduced)

image_pca_50 = b_pca_50_recovered[1,:].reshape([30, 40])
plt.imshow(image_pca_50)
plt.title('Compressed image with 50 components', fontsize=15, pad=15)
plt.savefig("image_pca_50.png")
```

```python
# Calculate the explained variance of 50 components
np.cumsum(pca_50.explained_variance_ratio_ * 100)[-1]
```

```python
pca_1000 = PCA(n_components=1000)
b_pca_1000_reduced = pca_1000.fit_transform(b_arr)
b_pca_1000_recovered = pca_1000.inverse_transform(b_pca_1000_reduced)

image_pca_1000 = b_pca_1000_recovered[1,:].reshape([30, 40])
plt.imshow(image_pca_1000)
plt.title('Compressed image with 1000 components', fontsize=15, pad=15)
plt.savefig("image_pca_1000.png")
```

```python
# Calculate the explained variance of 1000 components
np.cumsum(pca_1000.explained_variance_ratio_ * 100)[-1]
```
